```python
jb_skill_list = ["python","sql","ml","hf","nlp"]
```

```python
candidate_1  = ["python","excel"]
candidate_2  = ["python","data entry","ml"]
candidate_3 = ["python","excel","sql","hf"]
```

```python
candidate_skill_list= (candidate_1,candidate_2,candidate_3)
```

```python
def evaluate_candidates(jb_skill_list, candidate_skill_list):
    results = []
    
    for i, candidate in enumerate(candidate_skill_list, start=1):
        # Calculate missing skills
        missing_skills = [skill for skill in jb_skill_list if skill not in candidate]
        matched_skills = [skill for skill in jb_skill_list if skill in candidate]

        
        # Calculate match percentage
        matched_skills = [skill for skill in jb_skill_list if skill in candidate]
        if not matched_skills:
            percentage = 0.0
        else:
            percentage = round((len(matched_skills) / len(jb_skill_list)) * 100, 2)
        
        # Store results for this candidate
        results.append({
            "candidate_id": i,
            "jb_skill_list":jb_skill_list,
            "matched_skills":matched_skills,
            "missing_skills": missing_skills,
            "match_percentage": percentage
        })
    
    return results
```

```python
evaluate_candidates(jb_skill_list, candidate_skill_list)
```
