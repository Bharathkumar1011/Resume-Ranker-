```python
def generate_feedback(candidate_skill_list, jb_skill_list):
    feedback_reports = []
    
    for i, candidate_skills in enumerate(candidate_skill_list, 1):
        prompt = f"""
        Analyze ONLY the provided candidate data against the job description. 
        Do NOT invent any information not present in the data.
        If a category has no relevant data, state "No data available".

        Job Requirements: {", ".join(jb_skill_list)}

        Candidate Skills: {", ".join(candidate_skills) if candidate_skills else "No skills listed"}

        Generate feedback using ONLY this exact format and ONLY the provided data:

        Candidate ID: Candidate_{i}

        Analysis:
        ✓ Strengths (must be directly supported by scores/data):
        {generate_strengths(candidate_skills, jb_skill_list)}

        ✗ Missing Qualifications (must be directly supported by low scores):
        {generate_gaps(candidate_skills, jb_skill_list)}

        ↗ Recommendations (must derive directly from gaps):
        {generate_recommendations(candidate_skills, jb_skill_list)}
        """
        print(candidate_skills)
        response = client.chat.completions.create(
            model="llama3-70b-8192",
            messages=[
                {"role": "system", "content": "You are a data-driven career advisor. Use ONLY the provided data. Never invent information."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.3,
            max_tokens=500
        )
        
        feedback_reports.append(response.choices[0].message.content)

    return feedback_reports

def generate_strengths(candidate_skills, jb_skills):
    if not candidate_skills:
        return "- No sufficient data for strengths"
    
    matches = set(candidate_skills) & set(jb_skills)
    if not matches:
        return "- No direct matches with job requirements"
    
    strengths = []
    for skill in matches:
        strengths.append(f"- Strong match with required skill: {skill}")
    
    return "\n".join(strengths[:3])  # Return top 3 strengths

def generate_gaps(candidate_skills, jb_skills):
    if not candidate_skills:
        return "- No sufficient data for missing qualifications"
    
    missing = set(jb_skills) - set(candidate_skills)
    if not missing:
        return "- Candidate meets all technical requirements"
    
    gaps = []
    for skill in list(missing)[:3]:  # Top 3 gaps
        gaps.append(f"- Missing required skill: {skill}")
    
    return "\n".join(gaps)

def generate_recommendations(candidate_skills, jb_skills):
    if not candidate_skills:
        return "- No sufficient data for recommendations"
    
    missing = set(jb_skills) - set(candidate_skills)
    if not missing:
        return "- No additional technical skills needed"
    
    recommendations = []
    skill_priority = {
        "Python": "Python programming",
        "SQL": "SQL/database skills",
        "AWS": "cloud computing (AWS)"
    }
    
    for skill in list(missing)[:3]:  # Top 3 recommendations
        area = skill_priority.get(skill, skill)
        recommendations.append(f"- Develop {area} through courses or projects")
    
    return "\n".join(recommendations)
```
