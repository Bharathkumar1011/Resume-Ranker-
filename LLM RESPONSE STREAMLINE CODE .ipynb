```python
def generate_feedback(ranked_candidates, job_description):
    feedback_reports = []
    
    for candidate in ranked_candidates:
        prompt = f"""
        Analyze ONLY the provided candidate data against the job description. 
        Do NOT invent any information not present in the data.
        If a category has no relevant data, state "No data available".

        Job Requirements (truncated): {job_description[:500]}...

        Candidate Data:
        Name: {candidate['Name']}
        Total Score: {candidate['Total Score']:.2f}/1.0
        Skills: {candidate.get('Skills', 'No data')}
        Experience: {candidate.get('Experience', 'No data')}
        Education: {candidate.get('Education', 'No data')}

        Generate feedback using ONLY this exact format and ONLY the provided data:

        Candidate Name: {candidate['Name']}

        Analysis:
        ✓ Strengths (must be directly supported by scores/data):
        - [Specific strength from Skills/Experience/Education with evidence]
        - [Specific strength from Skills/Experience/Education with evidence]
        - [Specific strength from Skills/Experience/Education with evidence]

        ✗ Missing Qualifications (must be directly supported by low scores):
        - [Specific gap with evidence from low score/missing data]
        - [Specific gap with evidence from low score/missing data]
        - [Specific gap with evidence from low score/missing data]

        ↗ Recommendations (must derive directly from gaps):
        - [Specific actionable recommendation based on gaps]
        - [Specific actionable recommendation based on gaps]

        If any section cannot be filled from data, write:
        "No sufficient data for [section]"
        """

        response = client.chat.completions.create(
            model="llama3-70b-8192",
            messages=[
                {"role": "system", "content": "You are a data-driven career advisor. Use ONLY the provided data. Never invent information."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.3,  # Lower temperature for more factual responses
            max_tokens=500    # Limit response length
        )
        
        feedback_reports.append({
            "name": candidate["Name"],
            "score": candidate["Total Score"],
            "feedback": response.choices[0].message.content
        })
    
    return feedback_reports
```
